```python
"""
=============================================================
DBS Technologies Internship — Level 2
ASSESSMENT: Task 5
Topic: Data Analysis & Visualization
Format: Code Tasks + Chart Critique + Tool Selection
=============================================================

Instructions:
  - Complete ALL tasks.
  - Charts will be saved as PNG files in the same directory.
  - Written answers are stored as Python string variables.
  - Do not delete any existing code or task descriptions.
  - Download this task and upload to your github. Then share the link to the task.
  - Also Download the generated PNG files and upload along with your task solution.
=============================================================
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

YOUR_NAME = ""   # ← Ofordile Nkeiruka Victoria
YOUR_DATE = ""   # ← 22/04/2026

#print(f"Task 5 |  {"Victoria"}  |  {22/4/2026}")

# Shared dataset — used across multiple tasks
np.random.seed(42)
n = 80
df = pd.DataFrame({
    "region":    np.random.choice(["North", "South", "East", "West"], n),
    "product":   np.random.choice(["Laptop", "Phone", "Tablet", "Watch"], n),
    "revenue":   np.random.randint(500, 5000, n),
    "units_sold":np.random.randint(1, 50, n),
    "month":     np.random.choice(range(1, 12), n),
    "rating":    np.round(np.random.uniform(2.5, 5.0, n), 1),
})

# ─────────────────────────────────────────────────────────────
# PART A — MATPLOTLIB CHARTS
# ─────────────────────────────────────────────────────────────

# A1 — Multi-panel Figure
# Create a single figure with 2 subplots side-by-side:
#   Left:  Bar chart of total revenue per region
#          (each bar a different colour, values labelled on top)
#   Right: Line chart of average monthly revenue across all
#          regions (month 1–12 on x-axis, markers on each point)
# Add a descriptive title to the entire figure.
# Save as "A1_revenue_overview.png"

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Revenue Overview", fontsize=16, fontweight="bold")

# YOUR CODE HERE — left subplot (axes[0])
region_revenue = df.groupby("region")["revenue"].sum().sort_values(ascending=True)
colors = plt.cm.viridis(np.linspace(0, 1, len(region_revenue)))


axes[0].bar(region_revenue.index, region_revenue.values, color=colors)
axes[0].set_title("Total Revenue by Region")
axes[0].set_xlabel("Region")
axes[0].set_ylabel("Revenue")

# Add value labels on top
for i, v in enumerate(region_revenue.values):
    axes[0].text(i, v + 100, str(v), ha='center', fontsize=9)
        
# YOUR CODE HERE — right subplot (axes[1])
        # RIGHT: Line chart (average monthly revenue)
monthly_avg = df.groupby("month")["revenue"].mean().sort_index()

axes[1].plot(monthly_avg.index, monthly_avg.values, marker='o')
axes[1].set_title("Average Monthly Revenue")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Average Revenue")
axes[1].grid(True)

plt.tight_layout()
plt.savefig("A1_revenue_overview.png", dpi=150)
plt.show()
print("A1 saved.")


# A2 — Scatter Plot with Regression Line
# Plot units_sold (x-axis) vs revenue (y-axis).
# Colour each point by product category (4 colours, one per product).
# Add a single overall regression/trend line.
# Add a legend that shows which colour = which product.
# Save as "A2_units_vs_revenue.png"

# YOUR CODE HERE
plt.figure(figsize=(8,6))

sns.scatterplot(data=df, x="units_sold", y="revenue", hue="product")

# Regression line (no scatter, just line)
sns.regplot(data=df, x="units_sold", y="revenue", scatter=False)

plt.title("Units Sold vs Revenue")
plt.xlabel("Units Sold")
plt.ylabel("Revenue")

plt.legend(title="Product")
plt.savefig("A2_units_vs_revenue.png", dpi=150)
plt.show()

print("A2 saved.")


# A3 — Histogram with Stats Annotations
# Plot a histogram of the 'rating' column using 10 bins.
# On the same axes:
#   - Draw a vertical dashed line at the MEAN rating (red)
#   - Draw a vertical dashed line at the MEDIAN rating (green)
# Add a legend labelling the two lines with their values.
# Save as "A3_rating_distribution.png"

# YOUR CODE HERE
plt.figure(figsize=(8,6))

plt.hist(df["rating"], bins=10)

mean_rating = df["rating"].mean()
median_rating = df["rating"].median()

# Mean line
plt.axvline(mean_rating, linestyle='dashed', label=f"Mean: {mean_rating:.2f}")

# Median line
plt.axvline(median_rating, linestyle='dashed', label=f"Median: {median_rating:.2f}")

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.legend()
plt.savefig("A3_rating_distribution.png", dpi=150)
plt.show()

print("A3 saved.")
```
