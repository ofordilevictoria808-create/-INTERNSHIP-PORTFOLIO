```python
def is_prime(num):
    if num <= 1:
        return False
    for e in range(2, num):
        if num % e == 0:
                return False
        return True
                
print(is_prime(7)) 
            
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for e in range(2, num):
        if num % e == 0:
                return False
        return True
                
print(is_prime(20)) 
```

```python
def get_primes_up_to(n):
    primes = []
    for num in range(2, n + 1):
        if is_prime(num):
            #calls the first function
            primes.append(num)
    return primes
print(get_primes_up_to(20))  
            
```
