```python
import pandas as pd
import numpy as np
data = {
    "employee_id":  [101, 102, 103, 104, 105, 106, 107, 108],
    "department":   ["Engineering", "Marketing", "Engineering",
                     "HR", "Marketing", "Engineering", "HR", "Marketing"],
    "salary":       [85000, 52000, 91000, 47000, 61000, np.nan, 43000, 58000],
    "years_exp":    [5, 3, 8, 2, 4, 6, 1, 3],
    "performance":  ["High", "Medium", "High", "Low",
                     "Medium", "High", "Low", "Medium"]}
df = pd.DataFrame(data)

eng_mean = df[df["department"] == "Engineering"]["salary"].mean()

df.loc[(df["department"] == "Engineering") & (df["salary"].isna()),"salary"] = eng_mean
print(df)
```

```python
 # Storing the cleaned dataframe
C1_df = df.copy()
print(C1_df)
```

```python
def sort_salary(salary):
    if salary < 50000:
        return "Junior"
    if 50000 <= salary < 80000:
        return "Mid"
    else:
        return "Senior"
df["salary_band"] = df["salary"].apply(sort_salary)  
print(df)
```
